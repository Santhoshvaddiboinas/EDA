```python
values=df['marital-status'].value_counts()
```

```python
grouped = df.groupby(['marital-status', 'race']).size().reset_index(name='count')
races = df['race'].unique()
for race in races:
    subset = grouped[grouped['race'] == race]
    plt.figure(figsize=(5,5))
    plt.pie(subset['count'], labels=subset['marital-status'], autopct='%1.1f%%', startangle=140)
    plt.title(f'Marital Status Distribution for {race}')
    plt.show()
    print()
    print(end="")
```
